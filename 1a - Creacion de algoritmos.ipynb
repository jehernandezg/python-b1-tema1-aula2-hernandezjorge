# Algoritmos

## ¿Qué es un algoritmo?

Fácilmente, podemos entender un algoritmo como los pasos o instrucciones que hay que seguir para realizar una actividad o resolver cualquier problema planteado, tomando en cuenta que tienen que ser instrucciones precisas, que siguen un orden secuencial y son finitas.


### Algoritmos en la vida cotidiana
En la vida cotidiana, un algoritmo podría ser la rutina que realizamos para salir al trabajo, los pasos que se siguen para una receta de comida, los pasos que se realizan para arrancar un vehículo, entre otros ejemplos.

Como veremos ahora, un ejemplo de algoritmo muy común que encontramos cotidianamente son los pasos que seguimos para realizar un correcto lavado de manos.

image.png

Veamos ahora un ejemplo de algoritmo escrito en lenguaje natural.

_¿Qué pasos seguimos para encender la televisión?_
1. Buscamos el control remoto.
2. Nos paramos en frente del televisor.
3. Pulsamos el botón de encendido.


---


### Pasos para crear un algoritmo

Para crear un algoritmo debemos seguir una serie de pasos, que son los siguientes:

1. **Definición del problema:**
    _El problema planteado que se quiere solucionar._
2. **Análisis:** 
    _Qué datos de entrada vamos a proporcionar para solucionar el problema y qué datos deseamos obtener (resultado)._
3. **Diseño (pseudocódigo / diagramas de flujo):** 
    _Los pasos que vamos a seguir para solucionar._
4. **Implementación:** 
    _Desarrollar e implementar el código con todos los recursos anteriormente mencionados._
5. **Pruebas:** 
    _Las pruebas se realizan en las funciones que encapsulan nuestro algoritmo. Es considerada una buena práctica de programación y es muy importante para poder comprobar que la función cumple su objetivo. Existen distintas pruebas que se pueden desarrollar, como son las ```pruebas unitarias``` o las ```pruebas de integración```, que se explicarán próximamente, por lo que en este momento solo nos centraremos brevemente en las pruebas unitarias._ 
    
---

**EXTRA: ¿Cómo realizar pruebas unitarias?**
    
En los distintos lenguajes de programación existen las pruebas unitarias; sin embargo, en Python existen varias librerías, y entre las más destacadas encontramos ```pytest``` y ```unittest```. No obstante, se debe recordar que existen más librerías enfocadas a distintas pruebas.

El primer paso para probar nuestro código es desarrollar la función que encapsula nuestro algoritmo y que posteriormente comprobaremos.

Para realizar la prueba comenzaremos importando nuestra librería como vemos a continuación.


````python
import pytest
````

Dentro de esta librería tenemos ya funciones prediseñadas que nos ayudan, por ejemplo, a comparar si un elemento es igual, mayor o menor a otro, o si se encuentra un valor en una lista; de esta manera, podemos comprobar que el funcionamiento del código es correcto.

En ```pytest``` existe un patrón que se debe seguir en el momento de desarrollar el código, llamado AAA (_Arrange, Act and Assert_).

1. **Arrange / Organizar:** Definimos las condiciones que se comprobarán.
2. **Act / Actuar:** Ejecutamos el código a probar.
3. **Assert / Afirmar:** Verifica que el código al que se realizó la prueba funcione de la manera esperada.

**Nota:** Es recomendable que en el momento de nombrar la función que nos ayudará a probar nuestro código se la nombre con ```name_test``` o ```test_name```.

### Ejemplos

_¿Cómo encontramos el área de un rectángulo?_

Recordando los pasos anteriormente mencionados para crear un algoritmo, entonces:
1. **Definir el problema:** _El problema que se ha definido es hallar el área de un rectángulo._
2. **Análisis:** _Realizamos el análisis respectivo para hallar el área de un rectángulo, por lo que resolvemos multiplicar la base del rectángulo por su altura: `resultado = base * altura`._
3. **Diseño:** _Recordemos que los pseudocódigos son instrucciones precisas y concretas, y los diagramas de flujo son estructuras gráficas con símbolos definidos utilizados para representar algoritmos; ambos son normalmente considerados un paso previo a la implementación utilizando un lenguaje de programación como se muestra a continuación, respectivamente._

4. **Implementación:** _Una vez que tenemos desarrollado el pseudocódigo, al igual que el diagrama de flujo implementamos el código para lenguaje de Python._

````python
base = int(input("Ingrese la base: "))
altura =  int(input("Ingrese la altura: "))


def calcular_area(base,altura):
  return base * altura

resultado = calcular_area(base,altura)
print("El resultado del área de su rectángulo es: ", resultado)
````

5. **Pruebas:** _Por último, realizamos la prueba al código desarrollado mediante ```pytest```._

````python
import pytest
````

````python
#En este caso realizaremos la prueba al código realizado anteriormente.
#Arrange
#Definimos que testearemos la función llamada calcular_area.

#Act
def test_area (): 
  print("Test passed")
  #Enviamos a la función calcular_area los valores 4 y 5.
  assert calcular_area(4,5) == 20 #Igual al valor que obtendremos de respuesta.

#Assert
"""De acuerdo al resultado obtenido definiremos si hubo un AssertionError o si 
se comprobó que el código funcionaba de la manera esperada."""
test_area()
````
