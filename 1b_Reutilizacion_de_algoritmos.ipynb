# Reutilización de algoritmos

La reutilización de algoritmos consiste en la capacidad de utilizar código que ya ha sido escrito y probado en un programa para evitar tener que escribirlo de nuevo. Esto puede ayudar a ahorrar tiempo y esfuerzo, ya que no es necesario escribir y probar el código desde cero.

## Algoritmos disponibles en Python

Python cuenta con varias funciones estándar, es decir, funciones prediseñadas y disponibles para su uso en nuestros programas. Las funciones comúnmente utilizadas son:


Tipos de datos
* `int()`: Permite definir un dato como entero.
* `float()`: Permite definir un dato como decimal.
* `str()`: Permite definir un dato como una cadena de caracteres.
* `type()`: Permite conocer el tipo de dato de una variable.

Entrada y salida de datos
* `input()`: Permite que el usuario a través del teclado digite un dato de entrada.
* `print()`: Imprime mensajes o valores en pantalla.

Manejo de datos
* `len()`: Devuelve la longitud de una cadena de caracteres, listas, tuplas, etc.
* `max()`: Devuelve el valor máximo que se encuentra dentro de una secuencia.
* `min()`: Devuelve el valor mínimo que se encuentra dentro de una secuencia.
* `sorted()`: Ordena de menor a mayor una secuencia dada.


````python
# Demostración Tipo de datos
x = 1.0
x = int(x)
print(x, type(x))

x = float(x)
print(x, type(x))

x = str(x)
print(x, type(x))
````

````python
# Entrada y salida de datos
num = input("Introduzca un número: ")
print("El número ingresado es: ", num)
````

````python
# Manejo de datos
ejemplo = (9, 15, 7 ,2, 3, 50, 25)
````

````python
len(ejemplo)
````

````python
max(ejemplo)
````

````python
min(ejemplo)
````

````python
sorted(ejemplo)
````

### Uso de funciones

Las funciones son muy útiles para crear un código reutilizable debido a que son bloques de código que cumplen una tarea determinada y pueden ser llamadas reiteradas veces utilizando distintos datos de entrada.

La estructura a considerar al crear una función es la siguiente:

```def nombre_de_la_función(parámetros):```

De igual manera, se debe tener definida la tarea que debe realizar la función.

**Ejemplo**

_Función que realiza el cálculo de la hipotenusa de un triángulo rectángulo._

La fórmula para calcular la hipotenusa es ```c = sqrt(a^2 + b^2)```, por lo que nuestra función deberá realizar esa operación.

trian.jpg

````python
import math

def calcular_hipotenusa(a, b):
  c = math.sqrt(a**2 + b**2)
  return c

a = int(input("Introduzca el cateto 1: "))
b = int(input("Introduzca el cateto 2: "))

c = calcular_hipotenusa(a, b)
c = round(c,2)
print("El resultado de la hipotenusa es: ", c)
````

### Uso de paquetes

Al igual que las funciones, los paquetes o librerías nos sirven para reutilizar código, son carpetas que nos ayudan a organizar funciones y clases que mantengan tareas afines y necesiten ser agrupadas. Las librerías son una forma apropiada de distribuir y compartir código que puede ser reutilizado en diferentes proyectos.

Para poder crear un paquete debe crearse una carpeta y dentro de la misma crear un archivo que dé inicio a nuestro paquete.

```
mi_paquete/
  __init__.py
```

Dentro de la carpeta se pueden crear módulos y funciones, y el archivo debe contar con extensión ```.py```.

```
mi_paquete/
  operaciones.py
```
Se sigue la siguiente estructura para poder importar y ocupar lo que contenga el paquete.

```
import nombre_del_paquete
```

Estos conceptos se tratarán con más profundidad en el tema 3a (Diseño de software / modularidad en Python)

#### NumPy

Un paquete para el cálculo científico con Python que proporciona funcionalidades para trabajar con matrices y arrays.

Es utilizado en gestión de datos.

````python
import numpy as np

ejemplo = np.array([1, 2, 3, 4])
print(ejemplo)

ejemplo_2 = np.array([25, 26, 27, 28])
print(ejemplo_2)

print("\n Suma de dos arrays")
print(ejemplo + ejemplo_2)

print("\n Multiplicación de dos arrays")
print(ejemplo * ejemplo_2)

print("\n Multiplicación de un escalar por cada array")
print(5 * ejemplo)
print(2 * ejemplo_2)

print("\n Raíz de cada elemento del array")
print(np.sqrt(ejemplo_2))
````

#### Pandas

Permite el análisis y la manipulación de datos en gran escala, es decir, leerlos y escribirlos, y de igual manera permite mantener un orden en los datos.

La principal estructura que se encuentra en ```Pandas``` son los ```DataFrame```, que permiten tabular datos y los muestra en columnas y filas.

````python
import pandas as pd

#Diccionario a DataFrame
dic = [
    {'nombre': 'Daniel', 'edad': '25', 'país': 'Estados Unidos'},
    {'nombre': 'Jesse', 'edad': '23', 'país': 'Canadá'},
    {'nombre': 'Sam', 'edad': '30', 'país': 'Estados Unidos'},
    {'nombre': 'Dayan', 'edad': '20', 'país': 'España'}
]

dataframe = pd.DataFrame(dic)
print(dataframe)
````

#### Matplotlib

Permite crear gráficos estáticos e interactivos a partir de datos introducidos u obtenidos.

````python
import matplotlib.pyplot as plt

#Creamos los ejes para nuestro gráfico.
fig, ax = plt. subplots()

#Ocuparemos las variables de Numpy 'ejemplo' y 'ejemplo_2' respectivamente.
ax.plot(ejemplo, ejemplo_2)

# Título
plt.title("Matplotlib")

#Etiquetas de los ejes
plt.xlabel("Eje x")
plt.ylabel("Eje y")

#Mostramos la gráfica.
plt.show()
````
