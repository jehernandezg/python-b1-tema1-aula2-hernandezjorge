# Recursividad

## ¿Qué es la recursividad?

¿Te has preguntado qué sucedería si defines una función cualquiera en Python y dentro de la misma llamas nuevamente a la función?

En términos de programación, eso se conoce como recursividad (una función que se llama a sí misma dentro de la función). Normalmente, se usa en problemas que se puedan dividir en tareas más pequeñas y repetitivas.
Las llamadas de las funciones recursivas se almacenarán en memoria en forma de stack o pila, donde la primera instancia de la función en ser creada será la última en retornar.

Para entender la recursividad, podemos imaginar una muñeca "matrioska".

¿Cómo puedo saber cuántas muñecas existen?

Para ello definimos la acción de:
```
abrir_muñeca()
```

Ahora vamos a repetir la llamada a la función hasta que ya no se pueda abrir la muñeca; ese será nuestro caso base, que es donde se retorna un valor definido, y en este caso será 1.

Por lo tanto, podemos decir que en la primera llamada recursiva, el número total de muñecas es igual a: 1 muñeca + total de muñecas en el rectángulo.

Este proceso se repetirá hasta que se retorne el caso base.

_Llevemos la recursividad a la práctica._

### Diseño de una función recursiva

Para diseñar una función recursiva se tienen en cuenta los siguientes pasos:

1. **Definir el caso base**

Es la condición en la que nuestra llamada recursiva se termina. De manera similar sucede con los bucles de iteración, que tienen una condición donde se interrumpe el ciclo.

2. **Llamadas recursivas finitas**

Al ser una función que se llama a sí misma, podría fácilmente entrar en un bucle infinito o desbordamiento de la memoria, por lo tanto, es importante asegurarnos que la secuencia sea finita y que cada llamada se acerque más al caso base.

# Recursividad vs. iteración

¿Pero por qué usaríamos recursividad?

A continuación, tienes un listado de pros y contras de la recursividad vs. iteraciones.

## Pros

- El código con recursión es usualmente más corto, limpio y elegante que las iteraciones.
- Con recursión las tareas complejas se descomponen en tareas más simples.
- La generación de secuencias es más limpia con recursiones que con iteraciones.

## Contras

- El proceso de llamadas recursivas puede incrementar el uso de la memoria y la notación *Big O*, aunque existen herramientas como *memorización* (*Big O* y *memorización* se conocerán en sesiones posteriores).
- La lógica recursiva es más difícil de mantener y debuguear que las iteraciones.
- La recursión podría ser menos eficiente que la lógica iterativa.

## Factorial

Es el producto de todos los números enteros positivos iguales o menores a un número dado.

### Factorial con iteración

````python
def factorial_iterativo(numero: int):
    resultado = 1
    acumulador = 2
    while acumulador <= numero:
        resultado *= acumulador
        acumulador += 1
    return resultado

factorial_iterativo(3)
````

### Factorial con recursividad

````python
def factorial_recursivo(numero: int):
    if numero == 1: # Caso base
        return 1
    else:
        return numero * factorial_recursivo(numero-1) # El parámetro enviado se decrementa, por lo tanto, es finito.

factorial_recursivo(3)
````

A continuación podemos observar que la primera función llamada es la última en ser retornada.



```
factorial(3)          # 1ra llamada con 3
3 * factorial(2)      # 2da llamada con 2
3 * 2 * factorial(1)  # 3ra llamada con 1
3 * 2 * 1             # retorna desde 3ra llamada como número = 1
3 * 2                 # retorna desde 2da llamada
6                     # retorna desde 1ra llamada
```


## Serie de Fibonacci

Es una serie de números en la cual un número es la suma de los 2 anteriores.

```
[0, 1, 1, 2, 3, 5, 8, ...]
```

### Fibonacci con iteración


````python
def fibonacci_iterativo(numero: int):
    a, b = 0, 1
    for i in range(numero):
        a, b = b, a+b
    return a

fibonacci_iterativo(6)
````

### Fibonacci con recursividad

````python
def fibonacci_recursivo(numero: int):
    # En este algoritmo tenemos dos casos base, cuando el número es 0 o 1
    if numero == 0:
        return 0
    elif numero == 1:
        return 1
    else:
        return fibonacci_recursivo(numero-1) + fibonacci_recursivo(numero-2)

fibonacci_recursivo(6)
````

**Observación:** en el algoritmo de Fibonacci se pueden implementar métodos como *memorización* para evitar cálculos repetitivos y optimizar recursos computacionales.

### MCD (máximo común divisor)

Es el número por el que se pueden dividir dos o más números sin dejar un residuo.

### MCD con iteración

````python
def mcd_iterativo(a, b):
    for c in range(min(a, b), 0, -1):
        if a % c == b % c == 0:
            return c

mcd_iterativo(22, 77)
````

### MCD con recursividad

````python
def mcd_recursivo(a, b):
    if(b == 0):           # Condición del caso base
        return abs(a)
    else:
        return mcd_recursivo(b, a % b)

mcd_recursivo(22, 77)
````
